```python
#--->Operators: => Operators are used to perform operations on variables and values. Python has the following operators.
#1 : Arithmatic operators
#2: Comparison Operators
#3: Logical Operators
#4:  Bitwise Operators
#5:  Assignment Operators
#6: Identity Operators
#7: Membership Operators
```

```python
#1 : Arithmatic operators

a=5
b=5
c=a+b
print(c)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b)
print(a//b) # integer division
```

```python
#2: Comparison Operators
print(a==b)
print(a>b)
print(a<b)
print(a!=b)
```

```python
#3: Logical Operators

x=True
y=False
print(x or y)
```

```python
print(x and y)
```

```python
print( not y)
```

```python
#4:  Bitwise Operators
x=2
y=3
print(x & y)
```

```python
print(x | y)
```

```python
print(x >> 2)
```

```python
print(y << 3)
```

```python
#5:  Assignment Operators

a=3
print(a)
```

```python
a+=3  # a=a+3
print(a)
```
