```python
#------>Literals:=>
#Literals is a raw data given in a value. In Python, there are various types of literals.
#1. Numeric Literals
#2.String Literals
#3. Boolean Literals
#4. Special Literals
```

```python
#-->1. Numeric Literals

a=0b1010 #Binary Literals
b=100 #Decimal Literals
c=0o310 #Octal Literals
d=0x12c #Hexadecimal Literal

# Float Literals
float_1=10.5
float_2=1.5e2
float_3=1.5e-3

#Complex Literal

x = 3.14j
print(a, b,c,d)
print(float_1,float_2,float_3)
print(x,x.imag,x.real)
```

```python
#--->2.String Literals
string = 'This is Python'
strings =  "This is Python"
multiline_str = """Satyam"""

print(string )
print(strings )
print(multiline_str)
```

```python
#--->3. Boolean Literals

a = True +4
b = False + 10
print(a)
print(b)
```

```python
#--->#4. Special Literals
a = None
print(a)
```
