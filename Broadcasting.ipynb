```python
import numpy as np
```

```python
# Scenario 1
a1=np.arange(8).reshape(2,4)
a2=np.arange(8,16).reshape(2,4)
print(a1)
print(a2)
```

```python
a1+a2
```

```python
# Scenario 2
a3=np.arange(9).reshape(3,3)
a4=np.arange(3).reshape(1,3)
print(a3,a4)
```

```python
a3+a4 #this is Broadcasting
```

```python
#NumPy broadcasting is a mechanism that allows NumPy
#to perform operations on arrays of different shapes without
#explicitly creating copies of the smaller array to 
#match the larger array's shape. It provides a way to vectorize array 
#operations, making them more efficient and concise.   it is Broadcasting.
```

```python
#Rules of Broadcasting
#If x=m and y=n, operation will take place
a1=np.arange(8).reshape(2,4)
a2=np.arange(8,16).reshape(2,4)
a1+a2
```

```python
#if x=1 and y=n then also operation will take place(Same Dimension)
a3=np.arange(9).reshape(3,3)
a4=np.arange(3).reshape(1,3)
print(a3)
print(a4)
```

```python
a3+a4
```

```python
#If x=1 and y!=n then also operation will not take place
a9=np.arange(3).reshape(1,3)
a10=np.arange(16).reshape(4,4)
print(a9)
print(a10)
```

```python
a9+a10
```

```python
#If x=1 and n=1 then y==m operation to take place
a11=np.arange(3).reshape(1,3)
a12=np.arange(3).reshape(3,1)
print(a11)
print(a12)
```

```python
a11+a12
```

```python
#If x=1 and y=1 then the operation will take place no matter what
a13=np.arange(1).reshape(1,1)
a14=np.arange(20).reshape(4,5)
print(a13)
print(a14)
```

```python
a13+a14
```

```python
#If they are of different dimensions
a15=np.arange(4)
a16=np.arange(20).reshape(5,4)
print(a15)
print(a16)
```

```python
a15+a16
```
