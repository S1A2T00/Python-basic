```python
jackpot=random.randint(1,100)

guess=int(input("Please Guess now"))
counter=1

while guess!=jackpot:
    if guess < jackpot:
        print("guess higher")
    else:
        print("guess lower")
    guess=int(input("Please Guess now"))
    counter+=1

print("Sahi Jawab")
print("You took", counter,"attempts")
```
