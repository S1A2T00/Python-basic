```python
def mul(a,b):
    if(b==1):
        return a
    else:
        return a+mul(a,b-1)
print(mul(5,6))
```

```python
def fact(number):
    if number==1:
        return 1
    else:
        return number*fact(number-1)
print(fact(5))
```

```python
def palin(name):
    if len(name)==1:
        print("Palindrome")
    else:
        if name[0]==name[-1]:
            palin(name[1:-1])
        else:
            print("Not a palin")
palin("SATYAM")
palin("MADAM")
palin("malayalam")
palin("Malayalam")
        
```

```python
def fib(number):
    if number==0 or number==1:
        return 1
    else:
        return fib(number-1)+fib(number-2)
print(fib(12))


```

```python
import time
def fib(m):
    if m==0 or m==1:
        return 1
    else:
        return fib(m-1)+fib(m-2)
start=time.time()
print(fib(48))
print(time.time()-start)
```
